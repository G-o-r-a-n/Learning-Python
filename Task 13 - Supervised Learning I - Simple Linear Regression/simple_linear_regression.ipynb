# Insurance Charges Analysis

In this notebook, we will analyse how age affects insurance charges using a dataset of insurance charges. We'll import the required libraries, load the data, and take a quick look at its structure.

```python
# Import required libraries.
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Ensures plots are displayed within the notebook.
%matplotlib inline
```

```python
# Load the insurance dataset.
df = pd.read_csv("insurance.csv")
# Display first 5 rows of dataset to see column names.
df.head()
```

# Initial Data Exploration

To understand our dataset, we will plot a scatter plot of the 'age' and 'charges' variables. This will give us an initial idea of the relationship between these two variables.

```python
# Assign 'age' and 'charges' columns to 'age_data' and 'charges_data' respectively.
age_data = df[["age"]]
charges_data = df[["charges"]]
```

```python
# Create a scatter plot with 'age_data' on x-axis and 'charges_data' on y-axis.
plt.scatter(age_data, charges_data, color = "b")
# Set title and labels.
plt.title("Data: Insurance Charges vs. Age")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

# Model Training

We will use the Linear Regression model from the sklearn library to find a best-fit line (linear regression line) that can predict 'charges' based on 'age'. We'll then plot this line against our data to visualise the model's predictions.


```python
# Initialise a Linear Regression model.
charges_model = LinearRegression()
# Train model using our 'age' and 'charges' data.
charges_model.fit(age_data, charges_data)
```

```python
# Use the trained model to predict charges based on age.
predicted_charges = charges_model.predict(age_data)
# Plot the known datapoints (blue dots).
plt.scatter(age_data, charges_data, color = "b")
# Plot the linear regression line (red line) based on model's predictions.
plt.plot(age_data, predicted_charges, color = "r")
# Set title and labels.
plt.title("Data with Linear Regression Line: Insurance Charges vs. Age")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

# Predictions and Extended Line Plot

Next, we will use our trained model to predict the charges for ages up to 80 and plot an extended linear regression line. We will also make a specific prediction for the charges an 80-year-old individual might expect to pay.

```python
# Generate a range of ages up to 80 to test model.
extended_age_data = pd.DataFrame({'age': range(18, 81)})
# Predict charges for these ages to test model.
extended_predicted_charges = charges_model.predict(extended_age_data)
```

```python
# Plot the original scatter plot.
plt.scatter(age_data, charges_data, color = "b")
# Plot the extended linear regression line.
plt.plot(extended_age_data, extended_predicted_charges, color = "r")
# Set title and labels.
plt.title("Extended Linear Regression Line: Insurance vs. Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
# Prediction of 'charges' for an unknown datapoint.
print(
    f"Expected charges for an 80-year-old individual: ${extended_predicted_charges[-1][0]:,.2f}"
    )
```
